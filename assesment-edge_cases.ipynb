```python

def normalize_order_directions(client_orders):
    
    client_orders["Direction"] = client_orders["Direction"].map({"Buy": "B", "Sell": "S"})
    return client_orders

def compute_broker_summary(broker_group, broker_code, isin, matched_quantity, match_status):
    
    return {
        "Broker": broker_code,
        "ISIN": isin,
        "Total Matched Quantity": matched_quantity,
        "Total Brokerage": broker_group["Brokerage Amount"].sum(),
        "Total STT": broker_group["STT"].sum(),
        "Total Trades": len(broker_group),
        "Match Status": match_status
    }

def allocate_trade_quantities(client_group, broker_group, broker_code, isin, total_trade_quantity):
   
    matched_trades = []
    allocated_qty = 0  

    client_group = client_group.sort_values(by="Quantity", ascending=True)  

    for _, broker_trade in broker_group.iterrows():
        trade_qty = broker_trade["QTY"]
        brokerage = broker_trade["Brokerage Amount"]
        stt = broker_trade["STT"]
        trade_direction = broker_trade["Buy/Sell Flag"]
        cost_per_share = broker_trade["COST"]

        for _, client_order in client_group.iterrows():
            if client_order["Direction"] != trade_direction:
                continue  

            client_qty = client_order["Quantity"]
            matched_qty = min(client_qty, trade_qty)

            proportion = matched_qty / total_trade_quantity  
            allocated_brokerage = proportion * brokerage
            allocated_stt = proportion * stt
            execution_slippage = client_order["Quantity"] - matched_qty

            total_cost = (cost_per_share * matched_qty) + allocated_brokerage + allocated_stt

            matched_trades.append({
                "order_id": client_order["UCC"],  
                "trade_id": broker_trade.name,  
                "broker": broker_code,
                "symbol": client_order["Ticker"],
                "isin": isin,
                "matched_qty": matched_qty,
                "status": "Partial" if matched_qty < client_qty else "Matched",
                "execution_slippage": round(execution_slippage, 2),
                "allocated_brokerage": round(allocated_brokerage, 2),
                "allocated_stt": round(allocated_stt, 2),
                "total_cost": round(total_cost, 2)
            })

            trade_qty -= matched_qty
            allocated_qty += matched_qty

            if trade_qty == 0:
                break

    return matched_trades, allocated_qty

def process_trade_matching(client_orders, broker_trades):
    
    client_orders = normalize_order_directions(client_orders)
    
    matched_trades = [] 
    unmatched_trades = []
    broker_summary = []
    
    # client_orders  prespective
    
    for (isin, broker_code), broker_group in broker_trades.groupby(["Instrument ISIN", "party code/SEBI regn code of party"]):
        client_group = client_orders[client_orders["ISIN"] == isin]
        total_client_quantity = client_group["Quantity"].sum()
        total_trade_quantity = broker_group["QTY"].sum()

        if client_group.empty:
            broker_summary.append(compute_broker_summary(broker_group, broker_code, isin, 0, "Not Matched"))
            unmatched_trades.extend(broker_group.to_dict("records"))
            continue

        match_status = "Unmatched" if total_client_quantity > total_trade_quantity else "Fully Matched" if total_client_quantity == total_trade_quantity else "Excess"

        matched, allocated_qty = allocate_trade_quantities(client_group, broker_group, broker_code, isin, total_trade_quantity)
        matched_trades.extend(matched)

        if total_trade_quantity > total_client_quantity:
            unmatched_trades.append({
                "Trade_ID": broker_group.index[0],  
                "Broker": broker_code,
                "ISIN": isin,
                "Excess_QTY": total_trade_quantity - total_client_quantity,
                "Status": "Excess"
            })

        broker_summary.append(compute_broker_summary(broker_group, broker_code, isin, allocated_qty, match_status))

    return pd.DataFrame(matched_trades), pd.DataFrame(unmatched_trades), pd.DataFrame(broker_summary)


matched_trades_df, unmatched_trades_df, broker_summary_df = process_trade_matching(client_orders, broker_trades)
```

```python

def store_in_duckdb(matched_trades, unmatched_trades, broker_summary_df):
    conn = duckdb.connect("Update_trades_2.db")

    
    conn.execute("""
        CREATE TABLE IF NOT EXISTS reconciliation_client_results (
            order_id TEXT, trade_id INTEGER, broker TEXT, symbol TEXT, isin TEXT,
            matched_qty INTEGER, status TEXT, execution_slippage REAL,
            allocated_brokerage REAL, allocated_stt REAL, total_cost REAL
        )
    """)

    conn.execute("""
        CREATE TABLE IF NOT EXISTS unmatched_trades (
            Trade_ID INTEGER, Broker TEXT, ISIN TEXT, Excess_QTY INTEGER, Status TEXT
        )
    """)

    conn.execute("""
        CREATE TABLE IF NOT EXISTS broker_summary (
            Broker TEXT, ISIN TEXT, "Total Matched Quantity" INTEGER,
            "Total Brokerage" REAL, "Total STT" REAL, "Total Trades" INTEGER, "Match Status" TEXT
        )
    """)

    
    if unmatched_trades.empty:
        unmatched_trades = pd.DataFrame(columns=["Trade_ID", "Broker", "ISIN", "Excess_QTY", "Status"])

    if broker_summary_df.empty:
        broker_summary_df = pd.DataFrame(columns=["Broker", "ISIN", "Total Matched Quantity", "Total Brokerage", "Total STT", "Total Trades", "Match Status"])
    
    conn.register("matched_trades_temp", matched_trades)
    conn.register("unmatched_trades_temp", unmatched_trades)
    conn.register("broker_summary_df", broker_summary_df)

    
    if not matched_trades.empty:
        conn.execute("""
            INSERT INTO reconciliation_client_results 
            SELECT order_id, trade_id, broker, symbol, isin, matched_qty, status, 
                   execution_slippage, allocated_brokerage, allocated_stt, total_cost 
            FROM matched_trades_temp
        """)

    if not unmatched_trades.empty:
        conn.execute("""
            INSERT INTO unmatched_trades 
            SELECT Trade_ID, Broker, ISIN, Excess_QTY, Status 
            FROM unmatched_trades_temp
        """)

    if not broker_summary_df.empty:
        conn.execute("""
            INSERT INTO broker_summary 
            SELECT Broker, ISIN, "Total Matched Quantity", "Total Brokerage", 
                   "Total STT", "Total Trades", "Match Status" 
            FROM broker_summary_df
        """)

    conn.close()
    


store_in_duckdb(matched_trades_df, unmatched_trades_df, broker_summary_df)
```
